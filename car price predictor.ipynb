# encoding -> label encoding on object type data type

```python
da=dataset.select_dtypes(include="object")
for i in da:
    le=LabelEncoder()
    dataset[i]=le.fit_transform(dataset[i])
```

```python
# spliting the data in input and output
x=dataset.iloc[:,:-1] # input 
y=dataset["Selling_Price"] # output
```

```python
x=data
```
